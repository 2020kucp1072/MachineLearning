```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load the dataset

data = pd.read_csv('/content/Salary_Data.csv')
```

```python

# Display the first few rows of the dataset
print(data.head())
```

```python
# Extract the relevant columns: 'Humidity' (independent variable) and 'Apparent Temperature' (dependent variable)
X = data[['YearsExperience']].values  # Independent variable
Y = data['Salary'].values  # Dependent variable
```

Splitting the dataset

```python
# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Build and train the linear regression model
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
# Make predictions
Y_pred = model.predict(X_test)
```

```python

# Evaluate the model
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
```

```python


print(f"Mean Squared Error: {mse}")
print(f"R-squared Score: {r2}")
```

```python

# Visualize the results
plt.scatter(X_test, Y_test, color='blue', label='Actual Data')
plt.plot(X_test, Y_pred, color='red', label='Regression Line')
plt.xlabel('Humidity')
plt.ylabel('Apparent Temperature')
plt.title('Predicting Apparent Temperature from Humidity')
plt.legend()
plt.show()
```
